A linear boundary value problem is defined by,
y''=p(x)y'(x)+q(x)y(x)+r(x);a<=x<=b, with boundary conditions- y(a)=c,y(b)=d

```python
import sys
import os
sys.path.append(os.path.abspath("../Codes"))

from Linear_bvp.Shooting import lin_shooting
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

p="-2/x"
q="2/x**2"
r="sin(log(x))/x**2"
a=1
b=2
c=1
d=2
n=10
x_mesh,y_sol,yprime_sol,exact_soly,exact_primey=lin_shooting(p,q,r,a,b,c,d,n)
error_y=np.abs(exact_soly-y_sol)
error_yprime=np.abs(exact_primey-yprime_sol)
data=[[a,b,c,d,e,f,g] for a,b,c,d,e,f,g in zip(x_mesh,y_sol,exact_soly,error_y,yprime_sol,exact_primey,error_yprime)]
df=pd.DataFrame(data,columns=['x',"y(x)","Exact y","Error ||y-yexact||","y'(x)","Exact y'","Error ||y'-y'exact||"])
display(df)
```

Errors for y becomes less near the boundaries because the method is tuned to be like this.

```python
#visualization
plt.figure(figsize=(5,5))
plt.loglog(x_mesh,error_y,label="Error in y",color='blue',marker='x')
plt.loglog(x_mesh,error_yprime,label="Error in y'",color='orange',marker='^')
plt.title("Error vs x")
plt.xlabel("x")
plt.ylabel("Error")
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```
