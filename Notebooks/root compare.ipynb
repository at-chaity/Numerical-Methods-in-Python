```python
import sys
import os
sys.path.append(os.path.abspath("../Codes"))

from root_finding.bisection import bisection
from root_finding.newton_raphson import newton_raphson

# Run both methods
root_bi, iters_bi, hist_bi = bisection("x**3 - x - 2", a=1, b=2, tol=1e-6, n0=100, return_iters=True, return_history=True)
root_n, iters_n, hist_n = newton_raphson("x**3 - x - 2", p0=1, tol=1e-6, N0=100, return_iters=True, return_history=True)

# Print results
print("Root found by Bisection method =", root_bi)
print("Root found by Newton-Raphson method =", root_n)
print("Iterations needed for Bisection =", iters_bi)
print("Iterations needed for Newton-Raphson =", iters_n)
```

Newton-Raphson method calculates faster than bisection. Hence, it's memory efficient.

```python
import matplotlib.pyplot as plt

plt.plot(range(len(hist_bi)), hist_bi, marker='s', label="Bisection")
plt.plot(range(len(hist_n)), hist_n, marker='o', label="Newton-Raphson")
plt.title("Convergence Comparison")
plt.xlabel("Iteration")
plt.ylabel("Approximate Root")
plt.grid(True)
plt.legend()
plt.show()
```

Newton-Raphson method converged faster than Bisection.
